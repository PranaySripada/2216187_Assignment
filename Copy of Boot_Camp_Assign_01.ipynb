```python
#setss
s1={50,70,90,603,300}
s2={789,45,756}
print("s1=",s1)
print("s2=",s2)
s1.update(s2)
print("the set1(update)=",s1)
s2.add(11)
print("the set 2=",s2)
s2.remove(11)
print("the set 2=",s2)
s2.discard(45)
print("the set 2=",s2)
len(s1)
print("the length s1=",s1)
print("the intersection=",s1.intersection(s2))
print("the difference=",s1.difference(s2))

```
